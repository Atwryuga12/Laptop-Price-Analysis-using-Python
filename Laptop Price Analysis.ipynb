# Bivariate Analysis 

```python
data.info()
```

```python
plt.figure(figsize = (10, 8))
sns.boxplot(x = data['OS'], y = data['Price_euros'])
plt.title('OS vs CPU_frequency')
plt.show()
```

```python
plt.figure(figsize = (10, 8))
sns.barplot(x = data['Touchscreen'], y = data['Price_euros'], hue = data['Screen'])
plt.show()
```

```python
plt.figure(figsize = (10, 8))
sns.barplot(x = data['OS'], y = data['Price_euros'], hue = data['Touchscreen'])
plt.show()
```

```python
# plt.figure(figsize = (10, 8))
# sns.barplot(x = data['PrimaryStorage'], y = data['Price_euros'], hue = data['SecondaryStorage'])
# plt.show()

# Converting the 'SecondaryStorage' column to strings
data['SecondaryStorage'] = data['SecondaryStorage'].astype(str)


plt.figure(figsize=(10, 8))
sns.barplot(x=data['PrimaryStorage'], y=data['Price_euros'], hue=data['SecondaryStorage'])
plt.title("Barplot of Price by Primary and Secondary Storage")
plt.xlabel("Primary Storage")
plt.ylabel("Price (euros)")
plt.legend(title="Secondary Storage")
plt.show()
```
